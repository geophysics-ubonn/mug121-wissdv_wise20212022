```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

topography = pd.read_csv("mug121_psds_01_topography.csv", index_col=[0])
print(topography.head())
well = np.loadtxt("mug121_psds_01_well.txt")
print(well)
elec_model = pd.read_csv("mug121_psds_02_electrical_model.csv", index_col=[0])
print(elec_model.head())
pegel = pd.read_csv("mug121_psds_03_pegel.csv", parse_dates=True, index_col=["Zeitstempel"]).drop(["Zeitstempel.1"], axis=1)
print(pegel.head())
precipitation = pd.read_csv("mug121_psds_04_precipitation.csv", parse_dates=True, index_col=["Zeitstempel"]).drop(["Zeitstempel.1"], axis=1)
print(precipitation.head())
```

```python
well_depth = np.diff(well[:,1])
print(well_depth)
elec_layers = np.diff(elec_model["Tiefe z [m]"])
print(elec_layers)
pegel_min = pegel["Pegelstand [cm]"].min() / 100
pegel_max = pegel["Pegelstand [cm]"].max() / 100
print(pegel_min, pegel_max)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 5))
axes = ax.copy()

fig.suptitle("Projekt: Artesischer Brunnen")

axes[0,0].set_title("Topographie")
axes[0,0].plot(topography["GOK_x"], topography["GOK_z"], label="GOK")
axes[0,0].plot(well[:,0], well[:, 1], label="Brunnen")
signal = axes[0,0].plot(np.full(elec_model["Tiefe z [m]"].shape[0], well[0, 0]+10), elec_model["Tiefe z [m]"], label="elektrische Sondierung")
axes[0,0].set_xlabel("Profil x [m]")
axes[0,0].set_ylabel("Höhe/Tiefe z [m]")
axes[0,0].grid()
axes[0,0].legend()

axes[0,1].set_title("Geoelektrische Sondierung")
axes[0,1].semilogx(elec_model["spezifischer Widerstand rho [Ohmm]"], elec_model["Tiefe z [m]"], 
                   label=r"$\rho$ [$\Omega$m]")
axes[0,1].semilogx(elec_model["spezifische Leitfähigkeit sigma [S/m]"], elec_model["Tiefe z [m]"], 
                   label=r"$\sigma$ [$\frac{S}{m}$] bzw. $\frac{1}{\rho}$ [$\frac{1}{\Omega m}$]")
axes[0,1].set_xlabel("elektrisches Modell")
axes[0,1].set_ylabel("Tiefe z [m]")
axes[0,1].grid()
axes[0,1].legend()

axes[1,0].set_title("Pegelstand Brunnen")
axes[1,0].plot(pegel["Pegelstand [cm]"]/100, label=r"Pegelstand")
axes[1,0].set_xlabel("Zeit t [a]")
axes[1,0].set_ylabel("Pegelstand [m]")
axes[1,0].grid()
axes[1,0].legend()

axes[1,1].set_title("Niederschlag monatlich")
axes[1,1].plot(precipitation["Niederschlag [mm]"], label=r"Niederschlag")
axes[1,1].set_xlabel("Zeit t [a]")
axes[1,1].set_ylabel("Niederschlag [mm]")
axes[1,1].grid()

plt.tight_layout()
plt.savefig("mug121_artesischerBrunnen_all-datasets_NJH.jpg", bbox_inches="tight")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,4))

ax.set_title("Projekt: Artesischer Brunnen - Topographie und elektrisches Modell")

ax.semilogx(elec_model["spezifischer Widerstand rho [Ohmm]"], elec_model["Tiefe z [m]"], 
                   label=r"$\rho$ [$\Omega$m]", color="red")
ax.semilogx(elec_model["spezifische Leitfähigkeit sigma [S/m]"], elec_model["Tiefe z [m]"], 
                   label=r"$\sigma$ [$\frac{S}{m}$] bzw. $\frac{1}{\rho}$ [$\frac{1}{\Omega m}$]", color="orange")
ax.set_xlabel("elektrisches Modell [$\Omega$m] der geoelektrischen Sondierung")
# ax2.set_ylabel("Tiefe z [m]")
ax.grid()
ax.legend(loc="center right")

ax2 = ax.twiny()

ax2.plot(topography["GOK_x"], topography["GOK_z"], label="GOK", color="brown")
ax2.plot(well[:,0], well[:, 1], label="Brunnen", color="blue")
signal = ax2.plot(np.full(elec_model["Tiefe z [m]"].shape[0], well[0, 0]+10), elec_model["Tiefe z [m]"], label="elektrische Sondierung", color="green")
ax2.axhline(well[-1,1]+pegel_min, 0.45, 0.55, color="b", ls="--", label="min(Pegel)")
ax2.axhline(well[-1,1]+pegel_max, 0.45, 0.55, color="b", ls="-.", label="max(Pegel)")
ax2.set_xlabel("Profil x [m]", color="brown")
ax2.set_ylabel("Höhe/Tiefe z [m]")
ax2.legend(loc="center left")

plt.tight_layout()
plt.savefig("mug121_artesischerBrunnen_topo-elecsounding_NJH.jpg", bbox_inches="tight")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,4))

ax.set_title("Projekt: Artesischer Brunnen - Zeitreihen")
ax.plot(pegel["Pegelstand [cm]"]/100, label=r"Pegelstand", color="orange")
ax.plot(pegel["Pegelstand [cm]"].resample("12M").mean().rolling(window=10, min_periods=1, center=True).mean()/100, label=r"mittlerer Jahrespegelstand", ls="--", color="orange")
ax.axhline(np.abs(well[0,1]-well[-1,1]), color="k", ls="-.", label="Brunnenoberkante")
ax.axhline(np.abs(well[-1,1]-well[-1,1]), color="k", ls="--", label="Brunnenunterkante")
ax.axhline(pegel_min, color="b", ls="--", label="min(Pegel)")
ax.axhline(pegel_max, color="b", ls="-.", label="max(Pegel)")
ax.set_xlabel("Zeit t [a]")
ax.set_ylabel("Pegel [m]")
ax.legend(loc=2)

ax2 = ax.twinx()
ax2.plot(precipitation["Niederschlag [mm]"], label=r"Niederschlag", alpha=0.5)
ax2.plot(precipitation["Niederschlag [mm]"].resample("10a").mean().rolling(window=10, min_periods=1, center=True).mean(), color="b", ls="-", label=r"mittlerer Jahresniederschlag", )
ax2.set_ylabel("Niederschlag [mm]", color="b")
ax2.legend(loc=1)

plt.tight_layout()
plt.savefig("mug121_artesischerBrunnen_pegel-precipitation_NJH.jpg", bbox_inches="tight")
plt.show()
```
